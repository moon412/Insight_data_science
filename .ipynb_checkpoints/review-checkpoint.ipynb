```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
R2s_all = pd.read_csv('R2s_mean.csv', index_col=['subject'], na_values=['NAN'])
SUVR_all = pd.read_table('suvr.txt', index_col=['subject'])
```

```python
mask = ~(SUVR_all['amyloid'] == 'unknow')
```

```python
R2s_data = R2s_all.ix[mask, 'ctx-bankssts':'amygdala']
SUVR = SUVR_all.ix['AD4':'ADRCAG-99', :]
SUVR_data = SUVR.ix[mask, 'ctx-bankssts':'amygdala']
```

```python
print(R2s_data.shape)
print(SUVR_data.shape)
```

```python
paircorr = R2s_data.corrwith(SUVR_data)
plt.figure(figsize=(8, 18))
paircorr.plot.barh()
```

```python
figs, axes = plt.subplots(6, 7, sharex=False, sharey=True)
figs.set_size_inches(18, 18)
counter = 0
for row in range(6):
    for col in range(7):
        xdata, ydata = R2s_data.iloc[:, counter], SUVR_data.iloc[:, counter]
        r, p = scipy.stats.pearsonr(xdata, ydata)
        if r > 0.6 and p < 0.05:
            sns.regplot(xdata, ydata, color='r', ax=axes[row, col])
        else:
            sns.regplot(xdata, ydata, color='k', ax=axes[row, col])
        axes[row, col].set_xlabel('')
        axes[row, col].set_ylabel('')
        axes[row, col].set_xticklabels('')
        axes[row, col].set_title(R2s_data.columns[counter])
        axes[row, col].text(0.2, 0.8, 'r=%.2f\np=%.2f' % (r, p),
                           horizontalalignment='center',
                           verticalalignment='center',
                           transform=axes[row, col].transAxes)
        counter += 1
```
