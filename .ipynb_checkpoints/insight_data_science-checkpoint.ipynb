```python
import pandas as pd
import numpy as np
import scipy.stats 
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
%matplotlib inline
```

```python
R2s_AD = pd.read_csv('R2s_mean.csv', na_values=['NA'], index_col=['subject'])
R2s_AD.head(n=5)
```

```python
SUVR_AD = pd.read_table('suvr.txt', na_values=['NA'], index_col=['subject'])
SUVR_AD.head(n=5)
```

```python
mask = SUVR_AD['amyloid'].notnull()
R2s_data = R2s_AD.ix[mask, 'ctx-bankssts':'amygdala']
SUVR_data = SUVR_AD.ix[mask, 'ctx-bankssts':'amygdala']
print(SUVR_data.shape)
```

```python
paircorr = R2s_data.corrwith(SUVR_data)
plt.figure(figsize=(8, 16))
paircorr.plot.barh()
```

```python
SUVR_corr = SUVR_data.corrwith(SUVR_data['ctx-parahippocampal'])
plt.figure(figsize=(8, 16))
SUVR_corr.plot.barh()
```

```python
import seaborn as sns
import scipy.stats
import statsmodels.sandbox.stats.multicomp
```

```python
pearson = [scipy.stats.pearsonr(R2s_data[column], SUVR_data[column]) for column in R2s_data]
df_pearson = pd.DataFrame(pearson, columns=['r', 'p'], index=R2s_data.columns)
reject, pval_adjusted, alphaS, alphaB = statsmodels.sandbox.stats.multicomp.multipletests(df_pearson['p'], 
                                                                                          alpha=0.05, 
                                                                                          method='fdr_bh')
df_pearson['p_adjusted'] = pval_adjusted
df_pearson = df_pearson.T
df_pearson.head(n=3)
```

```python
figs, axes = plt.subplots(6, 7, sharex=False, sharey=True)
figs.set_size_inches(18, 18)
counter = 0
for i in range(6):
    for j in range(7):
        #r, p = scipy.stats.pearsonr(R2s_data.iloc[:, counter], SUVR_data.iloc[:, counter])
        r, p, ap = df_pearson.iloc[:,counter]
        xdata, ydata = R2s_data.iloc[:, counter], SUVR_data.iloc[:, counter]
        if r > 0.6 and ap < 0.05:
            sns.regplot(xdata, ydata, color='r', ax=axes[i, j])
        else:
            sns.regplot(xdata, ydata, color='b', ax=axes[i, j])
        axes[i, j].set_xticklabels([''])
        axes[i, j].set_title(R2s_data.columns[counter])
        axes[i, j].set_ylabel('')
        axes[i, j].set_xlabel('')
        axes[i, j].text(0.2, 0.8, 'r=%.2f\np=%.2f\nap=%.2f' % (r, p, ap), 
                        horizontalalignment='center', 
                        verticalalignment='center',
                        transform=axes[i, j].transAxes)
        counter += 1
```

```python
def region_correction(df):
    bad_regions = ['ctx-entorhinal','ctx-frontalpole','ctx-inferiortemporal','ctx-lateralorbitofrontal',
   'ctx-medialorbitofrontal','ctx-pericalcarine','ctx-temporalpole','ctx-transversetemporal']
    return df.drop(bad_regions, axis=1)    
```

```python
R2s_ctx = R2s_AD.loc[:, 'ab42_stat':'ctx-transversetemporal']
R2s_ctx = region_correction(R2s_ctx.dropna())
```

```python
from sklearn.decomposition import PCA
pca = PCA(whiten=False)
R2s_ctx_y = R2s_ctx['ab42_stat']
R2s_ctx_x = R2s_ctx.loc[:, 'ctx-bankssts':'ctx-supramarginal']
pca.fit(R2s_ctx_x)
R2s_ctx_pca = pca.transform(R2s_ctx_x)
```

```python
plt.figure()
plt.clf()
plt.scatter(R2s_ctx_pca[:, 0], R2s_ctx_pca[:, 1], c=R2s_ctx_y, s=30)
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib qt
matplotlib.style.use('classic')
fig = plt.figure(figsize=(6, 4))
plt.clf()
ax_pca = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=48, azim=134)
ax_pca.scatter(R2s_ctx_pca[:, 1], R2s_ctx_pca[:, 2], R2s_ctx_pca[:, 3], c=R2s_ctx_y)
```

```python
df_pca_comp = pd.DataFrame(pca.components_, columns=R2s_ctx_x.columns)
df_pca_comp.shape
```

```python
%matplotlib inline
plt.figure(figsize=(6, 4))
plt.plot(pca.explained_variance_ratio_)
```

```python
def biplot(pca, x_trans, y, labels):
    #pca is returned from sklearn.decomposition.PCA()
    #x_trans = pca(x)
    #y is a Series of 
    first_comp, second_comp = x_trans[:, 0], x_trans[:, 1]
    first_rot, second_rot = pca.components_[0, :], pca.components_[1, :]
    plt.figure(figsize=(12, 8))
    plt.scatter(first_comp, second_comp, c=y, s= 50)
    for region in range(len(first_rot)):
        plt.arrow(0, 0, first_rot[region]*max(first_comp)*1.2, 
                  second_rot[region]*max(second_comp)*1.2, color='b')
        plt.text(first_rot[region]*max(first_comp)*1.2*1.1, 
                 second_rot[region]*max(second_comp)*1.2*1.1,
                 labels[region], color='b')    
```

```python
biplot(pca, R2s_ctx_pca, R2s_ctx_y, R2s_ctx_x.columns)
```

```python
pca_SUVR = PCA()
SUVR_ctx_x, SUVR_ctx_y = SUVR_AD.loc[mask, 'ctx-bankssts':'ctx-transversetemporal'], SUVR_AD.loc[mask, 'ab42_stat']
#print([SUVR_ctx_x.shape, SUVR_ctx_y.shape])
pca_SUVR.fit(SUVR_ctx_x)
SUVR_ctx_pca = pca_SUVR.transform(SUVR_ctx_x)
```

```python
biplot(pca_SUVR, SUVR_ctx_pca, SUVR_ctx_y, SUVR_ctx_x.columns)
```
